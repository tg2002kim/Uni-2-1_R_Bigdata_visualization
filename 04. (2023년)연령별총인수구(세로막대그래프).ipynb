# 4) 데이터 전처리하기

```python
# Col 레이블 바꾸기
df.columns = ['행정구역', '9세 이하', '10대', '20대', '30대', '40대', '50대', '60대', '70대', '80대', '90대', '100세 이상']
df
```

```python
# 행정구역을 공백(' ')을 기준으로 분리하고 공백 왼쪽의 행정구역명만 남긴다.
# str.split : 문자열을 분리한다.
# expand=True : DataFrame 구조의 변경 사항을 적용한다.
# get(0) : 공백을 기준으로 분리된 0번째 데이터만 남긴다.
# 남겨진 행정구역을 행정구역 열에 배치한다.
df['행정구역'] = df['행정구역'].str.split(' ', expand=True).get(0)
df
```

```python
# index를 '행정구역'으로 바꾸기
# inplace=True : 원본 데이터 바꾸기
df.set_index('행정구역', inplace=True)
df
```

```python
# '전국' row 삭제하기
df.drop(index = '전국', inplace=True)
df
```

```python
# 전체 Row 중에서 NaN가 포함된 Row만 삭제하기
df.dropna(inplace=True)
df
```

```python
# 행 개수 정보 알아내기
df.shape
```

```python
# 문자(str) 콤마(,)를 제거해 정수형(int)으로 바꾸기(예) 1,951(문자) -> 1951(정수))
# 재 실행시에는 데이터 불러오기 후 다시 실행해야 한다.

# for 반복문 : i 변수의 값이 0~17까지 총 19회 반복하여 문자를 정수로 형변환한다.
# df.iloc[i] : integerLocation의 index
# str.replace(',', '').astype(int) : 문자열(str)에서 콤마(,)를 빈 칸으로 바꾸어, 정수형(int)로 형변환한다.

for i in range(17):
    df.iloc[i] = df.iloc[i].str.replace(',', '').astype(int)
df
```

## 5-2) 세로 막대 그래프 꾸미기

* 색상 맵 : https://matplotlib.org/stable/users/explain/colors/colors.html
* 패턴(hatch) 스타일 : https://matplotlib.org/stable/gallery/shapes_and_collections/hatch_style_reference.html

```python
# 막대 색상 지정하기

plt.figure(figsize=(10,5))                           # 그래프 크기(너비, 높이) 변경하기
area = '서울특별시'

# 세로 막대 그래프(x(연령대), y(특정지역의 인구수))
# color : 막대 색상
# alpha : 투명도(0~1)
plt.bar(df.columns, df.loc[area]//1000, color='green', alpha=0.7)

plt.title(area + ' 연령대별 인구수 비교(단위 : 천 명)')    # 그래프 제목
plt.xlabel('연령대')                                # x축 제목
plt.ylabel('인구수(천 명)')                         # y축 제목
plt.show()                                          # 그래프만 표시하기
```

```python
# 세로 막대 패턴(hatch) 채우기

plt.figure(figsize=(10,5))                           # 그래프 크기(너비, 높이) 변경하기
area = '서울특별시'

# 세로 막대 그래프(x(연령대), y(특정지역의 인구수))의 테이블을 bar 변수에 할당한다.
bar = plt.bar(df.columns, df.loc[area]//1000, color='green', alpha=0.7)
# index 5번째 '50대'에 점(..) 패턴(hatch) 스타일을 적용한다.
# '/' 사선, 'x'등 위의 패턴(hatch) 스타일 링크 참조
bar[5].set_hatch('..')

plt.title(area + ' 연령대별 인구수 비교(단위 : 천 명)')    # 그래프 제목
plt.xlabel('연령대')                                # x축 제목
plt.ylabel('인구수(천 명)')                         # y축 제목
plt.show()                                          # 그래프만 표시하기
```

```python
# 세로 막대 두께, 항목 레이블 기울기 변경하기

plt.figure(figsize=(10,5))                           # 그래프 크기(너비, 높이) 변경하기
area = '서울특별시'

# width : 막대 두께
bar = plt.bar(df.columns, df.loc[area]//1000, color='green', width=0.5)
bar[5].set_hatch('..')

plt.title(area + ' 연령대별 인구수 비교(단위 : 천 명)')    # 그래프 제목
plt.xlabel('연령대')                                # x축 제목
plt.ylabel('인구수(천 명)')                         # y축 제목

# 항목 개수가 많을 때 항목 레이블 기울게 표현한다.
plt.xticks(rotation = 45)       # x 축 레이블의 각도를 45도 회전
plt.yticks(rotation = 45)       # y 축 레이블의 각도를 45도 회전

plt.show()                                          # 그래프만 표시하기
```
