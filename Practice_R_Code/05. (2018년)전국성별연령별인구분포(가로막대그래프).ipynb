```python
# 구글 드라이브에서 외부 데이터 파일 불러오기

# 1. 구글 드라이브 마운트하기 : [파일-[드라이브 마운트] 아이콘
# 2. 구글 드라이브 파일 경로 복사하기 : content\drive\MyDrive\..\파일명 위에서 우 클릭 [경로 복사]
# 3. 구글 드라이브에서 외부 데이터 파일 불러와 file_name 변수에 할당하기
file_path = '/content/drive/MyDrive/파이썬데이터시각화/SpaceProbe_data/연령별인구현황(201812_202312).csv'

# csv 파일에서 행정구역, 년도별_총인구수 데이터 가져오기
# pd.read_csv('\경로\파일명.csv')
# encoding : csv를 불러올때 한글이 한글이 깨져 열릴때 'cp949'나 'utf-8'로 불러오기
# index_col = index or 'column 명' : row index로 사용할 Col 설정
# nrows = 1 : 1 row까지만 불러오기(header 포함)
# usecols=['월', '품목', '수량'] : 가져올 col 설정 / xlsx 파일에서는 열 머리글로 설정 uscols=('A', 'B:F')
# header = True(default) or False : 첫 줄을 Column 명으로 사용 여부
# sep = ',' : Col 구분자를 설정한다.
# skiprows = 3 또는 [0, 2]) : 첫 번째와 세 번째 row를 제외하고 불러옴(header 포함)
# names = [col_1, col_2, ....] : 칼럼명을 col_1, col_2...로 변경
# na_values =['a'] : NaN을 특정 값(a)로 변경
```

```python
# 2018년 전국 남자 연령대별 인구수 가져오기------------------------------------------------------------------------------------------
# index는 '행정구역'이고, 첫 번째 머리글을 제외한 하나 row(전국) 가져오기
df_m = pd.read_csv(file_path, encoding ='CP949', index_col='행정구역', nrows = 1, usecols=['행정구역','2018년_남_0~9세', '2018년_남_10~19세', '2018년_남_20~29세', '2018년_남_30~39세', '2018년_남_40~49세', '2018년_남_50~59세', '2018년_남_60~69세', '2018년_남_70~79세', '2018년_남_80~89세', '2018년_남_90~99세', '2018년_남_100세 이상'])
df_m
```

```python
# 2018년 전국 여자 연령대별 인구수 가져오기------------------------------------------------------------------------------------------
# index는 '행정구역'이고, 첫 번째 머리글을 제외한 하나 row(전국) 가져오기
df_f = pd.read_csv(file_path, encoding ='CP949', index_col='행정구역', nrows = 1, usecols=['행정구역', '2018년_여_0~9세', '2018년_여_10~19세', '2018년_여_20~29세', '2018년_여_30~39세', '2018년_여_40~49세', '2018년_여_50~59세', '2018년_여_60~69세', '2018년_여_70~79세', '2018년_여_80~89세', '2018년_여_90~99세', '2018년_여_100세 이상']
)
df_f
```

# 4) 데이터 전처리하기

```python
# 남자 index 값 바꾸기
# '전국  (0000000000)'를 '전국(남)'로 바꾸기
# inplace=True : 원본 값 바꾸기
df_m.rename(index={'전국  (0000000000)':'전국(남)'}, inplace=True)
df_m

# 여자 index 값은 하나의 DataFrame으로 합칠 것이므로 지금은 바꿀 필요가 없다.
```

```python
# 남자 Col 레이블 바꾸기
df_m.columns = ['9세 이하', '10대', '20대', '30대', '40대', '50대', '60대', '70대', '80대', '90대', '100세 이상']
df_m
```

```python
# 여자 Col 레이블 바꾸기
df_f.columns = df_m.columns
df_f
```

```python
# 남자 DataFrame의 마지막 row에 여자 DataFram의 1row 값을 추가하기
# df_m.loc['전국(여)'] : 남자 DataFrame의 마지막 row에 '전국(여)'라는 index value로 추가된다.
# df_f.iloc[0] : 여자 DataFrame의 1row
df_m.loc['전국(여)'] = df_f.iloc[0]
# 'df_m' DataFrame에 '전국(여)' row가 추가된다.
df = df_m       # 'df_m' DataFrame을 'df' 이름의 DataFrame에 할당한다.
df              # 'df' DataFrame 출력하기
```

```python
# 문자(str) 콤마(,)를 제거해 정수형(int)으로 바꾸기(예) 1,951(문자) -> 1951(정수))
# 재 실행시에는 데이터 불러오기 후 다시 실행해야 한다.

# for 반복문 : i 변수의 값이 0~1까지 총 2회 반복하여 문자를 정수로 형변환한다.
# df.iloc[i] : integerLocation의 index
# str.replace(',', '').astype(int) : 문자열(str)에서 콤마(,)를 빈 칸으로 바꾸어, 정수형(int)로 형변환한다.

for i in range(2):
    df.iloc[i] = df.iloc[i].str.replace(',', '').astype(int)
df
```

## 5-1) 가로 막대 그래프 만들기

* plt.barh : 가로 막대 그래프

```python
plt.figure(figsize=(10,6))                           # 그래프 크기(너비, 높이) 변경하기

# 남자 데이터 시각화----------------------------------------------------------------------------------
plt.barh(df.columns, df.iloc[0]//1000, label='남자')

plt.title('2018년 전국 남자 성별 인구 분포')    # 그래프 제목
plt.xlabel('인구수(천 명)')                     # x축 제목
plt.ylabel('연령대')                            # y축 제목
plt.show()                                      # 그래프만 표시하기
```

```python
plt.figure(figsize=(10,6))                           # 그래프 크기(너비, 높이) 변경하기

# 남자 데이터 시각화----------------------------------------------------------------------------------
plt.barh(df.columns, df.iloc[0]//1000, label='남자')
# 여자 데이터 시각화----------------------------------------------------------------------------------
plt.barh(df.columns, df.iloc[1]//1000, label='여자')

plt.title('2018년 전국 남녀 인구 분포')         # 그래프 제목
plt.xlabel('인구수(천 명)')                     # x축 제목
plt.ylabel('연령대')                            # y축 제목
plt.show()                                      # 그래프만 표시하기
```

```python
plt.figure(figsize=(10,6))                           # 그래프 크기(너비, 높이) 변경하기

# 남자 데이터 시각화----------------------------------------------------------------------------------
plt.barh(df.columns, -df.iloc[0]//1000, label='남자')   # 남자 데이터를 음수(-)로 처리한다.
# 여자 데이터 시각화----------------------------------------------------------------------------------
plt.barh(df.columns, df.iloc[1]//1000, label='여자')

plt.title('2018년 전국 남녀 인구 분포')         # 그래프 제목
plt.xlabel('인구수(천 명)')                     # x축 제목
plt.ylabel('연령대')                            # y축 제목
plt.show()                                      # 그래프만 표시하기
```

## 5-2) 가로 막대 그래프 꾸미기

* 범례 위치 :

```python
plt.figure(figsize=(10,6))                           # 그래프 크기(너비, 높이) 변경하기

# 남자 데이터 시각화----------------------------------------------------------------------------------
plt.barh(df.columns, -df.iloc[0]//1000, label='남자')
# 여자 데이터 시각화----------------------------------------------------------------------------------
plt.barh(df.columns, df.iloc[1]//1000, label='여자')

plt.title('2018년 전국 남녀 인구 분포', size=20)     # 그래프 제목
plt.xlabel('인구수(천 명)')                          # x축 제목
plt.ylabel('연령대')                                 # y축 제목

# 범례 추가하기-------------------------------------------------------------------------
# label + legend()
plt.legend()

# ncols=2 : 열 2개로 표시
# fontsize=10 : 범례 글꼴 크기
# shadow=True : 그림자
# plt.legend(ncols=2, fontsize=10, shadow=True)
plt.show()                                      # 그래프만 표시하기
```
